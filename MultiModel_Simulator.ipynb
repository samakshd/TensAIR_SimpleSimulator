#Processing Data Batch 'Blue'

```python
X_blue, y_blue = data_blue.iloc[:,:-1], data_blue.iloc[:,-1]
```

###Normalizing the data

```python
from sklearn.preprocessing import MinMaxScaler

# Scale the dataset
X_blue_scale = MinMaxScaler().fit(X_blue)
X_blue_trans = X_blue_scale.transform(X_blue) # fit and transform the data
X_blue = pd.DataFrame(X_blue_trans, columns = list(X_blue.columns)) # convert matrix to data frame with columns

y_blue_scale = MinMaxScaler().fit(np.array(y_blue).reshape(-1, 1))
y_blue = y_blue_scale.transform(np.array(y_blue).reshape(-1, 1))
```

###Running SGD on given data

```python
X_blue = np.column_stack(([1]*X_blue.shape[0], X_blue)) # add a column with ones for the bias value while converting it into a matrix
m,n = X_blue.shape # rows and columns 

theta_blue = [1]*n # initial theta
gradients_blue = [0]*n

X_blue = np.array(X_blue) # convert X into a numpy matrix
y_blue = y_blue.flatten() # convert y into an array

alpha_blue = 0.001 # alpha value 
iteration_blue = 100 # iterations
cost_blue = [] # list to store cost values
theta_new_blue = [] # list to store updates coeffient values
```

```python
for i in range(0, iteration_blue):
    pred = np.matmul(X_blue,theta_blue) # Calculate predicted value
    J = 1/2 * ((np.square(pred - y_blue)).mean()) # Calculate cost function
   
    t_cols = 0 # iteration for theta values
    
    # Update the theta values for all the features with the gradient of the cost function
    for t_cols in range(0,n): 
        t = round(theta_blue[t_cols] - alpha_blue/m * sum((pred-y_blue)*X_blue[:,t_cols]),4) # calculate new theta value
        gradients_blue[t_cols] += (alpha_blue/m * sum((pred-y_blue)*X_blue[:,t_cols]))*1
        theta_new_blue.append(t) # save new theta values in a temporary array
        
# update theta array
    theta_blue = [] # empty the theta array
    theta_blue = theta_new_blue # assign new values of theta to array
    theta_new_blue = [] # empty temporary array
    cost_blue.append(J) # append cost function to the cost array
```
